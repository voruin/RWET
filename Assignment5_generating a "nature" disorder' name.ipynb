```python
from transformers import pipeline, AutoModelForCausalLM, AutoTokenizer
import torch
import random
import markovify
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv('disorders.csv', delimiter='\t')
```

```python
df
```

```python
disorders_data = df['Disorder, condition, or problem'].values
disorders_data_unique=[]
# for p in disorders_data:
#     wordlist=p.split(",")
#     for w in wordlist:
#         if w.lower() not in disorders_data_unique:
#             disorders_data_unique.append(w.lower())
for p in disorders_data:
    word=p.split(",")[0]
    if word.lower() not in disorders_data_unique:
        disorders_data_unique.append(word.lower())

disorders_text = "\n".join(disorders_data_unique)
print(disorders_text)
```

```python
class LinesByChar(markovify.NewlineText):
    def word_split(self, sentence):
        return list(sentence)
    def word_join(self, words):
        return "".join(words)
```

```python
disorders_char_model = LinesByChar(disorders_text, state_size=4)
```

```python
for i in range(10):
    print(disorders_char_model.make_sentence(test_output=False).capitalize())
```

```python
disorders_char_model_phrase = markovify.NewlineText(disorders_text, state_size=4)
for i in range(10):
    print(disorders_char_model_phrase.make_sentence(test_output=False))
```

```python
import re

nature_data=open('nature2.txt').read()
# nature_data=re.findall(r'\b[A-Z]?[a-z]+\b', nature_data)
nature_data=nature_data.split("\n")
nature_data_unique=[]

for w in nature_data:
    if w.lower() not in nature_data_unique:
        nature_data_unique.append(w.lower())
        
nature_text="\n".join(nature_data_unique)
print(nature_text)
```

```python
nature_char_model = LinesByChar(nature_text, state_size=4)
```

```python
for i in range(10):
    print(nature_char_model.make_sentence(test_output=False))
```

```python
nature_char_model_phrase = markovify.NewlineText(nature_text, state_size=4)
for i in range(10):
    print(nature_char_model_phrase.make_sentence(test_output=False))
```

```python
output_n = 10
weights = [0.5,0.5]
length_limit = 50

combo_model=markovify.combine([disorders_char_model,nature_char_model],weights)

for i in range(output_n):
    out = combo_model.make_sentence(max_words=length_limit, test_output=False)
    out = out.replace("\n", " ")
    print(out)
    print()
```
