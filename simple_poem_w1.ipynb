```python
import random
import datetime


def generate_random_time_sentence():
    hours = [i for i in range(0, 12)]
    minutes = [i for i in range(0, 60)]
    random_hour = str(random.choice(hours))
    random_minute = str(random.choice(minutes))
#     if len(random_hour) == 1:
#         random_hour = "0" + random_hour
    if len(random_minute) == 1:
        random_minute = "0" + random_minute
    time = random_hour + ":" + random_minute
    sentence = "at " + time + " in the "+ random.sample(['morning', 'afternoon', 'evening'], k=3)[0]
    return sentence
```

```python
def generate_random_complete_sentence():
    p = inflect.engine()
    chooseTense=['past','present','continuous']
    chooseNumber=['singular', 'plural']
    chooseVerbs=['verbT','verbI']
    singularWords=['a','the','my','your']
    subject=''
    predicate=''
    object_b=''
    #choose subject
    number1=random.choice(chooseNumber)
    tense1=random.choice(chooseTense)
    number2=random.choice(chooseNumber)
    if(number1=='singular'):
        subject+=random.choice(singularWords)+" "
        if (random.random()<1/3):
            subject+=random.choice(adjectives_ahead)+" "
            subject+=random.choice(nouns)+" "
        elif (random.random()>=2/3):
            subject+=random.choice(nouns)+" "
            subject+=random.choice(adjectives_after)+" "
            subject+=p.plural(random.choice(nouns))+" "
        else:
            subject+=random.choice(nouns)+" "
    else:
        if (random.random()<1/3):
            subject+=random.choice(adjectives_ahead)+" "
            subject+=p.plural(random.choice(nouns))+" "
        elif (random.random()>=2/3):
            subject+=p.plural(random.choice(nouns))+" "
            subject+=random.choice(adjectives_after)+" "
            subject+=p.plural(random.choice(nouns))+" "
        else:
            subject+=p.plural(random.choice(nouns))+" "
    
    #choose predicate
    if(tense1!='continuous'):
        if (random.choice(chooseVerbs)=='verbT'):
            predicate+=conjugate(random.choice(verbTs).split()[0], tense = tense1, number = number1)+" "
        elif (random.choice(chooseVerbs)=='verbI'):
            predicate+=conjugate(random.choice(verbIs).split()[0], tense = tense1, number = number1)+" "
            predicate+=random.choice(preps)+" "
    else:
        if (number1=='singular'):
            predicate+="is"+" "
        else: 
            predicate+="are"+" "
        
        if (random.choice(chooseVerbs)=='verbT'):
            predicate+=lexeme(random.choice(verbTs).split()[0])[2]+" "
        elif (random.choice(chooseVerbs)=='verbI'):
            predicate+=lexeme(random.choice(verbIs).split()[0])[2]+" "
            predicate+=random.choice(preps)+" "

    #choose object
    if(number2=='singular'):
        object_b+=random.choice(singularWords)+" "
        if (random.random()<1/3):
            object_b+=random.choice(adjectives_ahead)+" "
            object_b+=random.choice(nouns)
        elif (random.random()>=2/3):
            object_b+=random.choice(nouns)+" "
            object_b+=random.choice(adjectives_after)+" "
            object_b+=p.plural(random.choice(nouns))
        else:
            object_b+=random.choice(nouns)
    else:
        if (random.random()<1/3):
            object_b+=random.choice(adjectives_ahead)+" "
            object_b+=p.plural(random.choice(nouns))
        elif (random.random()>=2/3):
            object_b+=p.plural(random.choice(nouns))+" "
            object_b+=random.choice(adjectives_after)+" "
            object_b+=p.plural(random.choice(nouns))
        else:
            object_b+=p.plural(random.choice(nouns))
    return subject+predicate+object_b
```
